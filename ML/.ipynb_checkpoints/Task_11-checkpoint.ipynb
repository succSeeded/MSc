Now, let's implement the nnls linreg for the 2D case. Here I decided to convert it to make the nnls a constrained optimization problem. I used a penalty function that drastically increases the loss outside of the search region without affecting it on the inside.

```python
class nnls_custom:

    def __init__(self):

        self.weights_ = None

    def fit(self, x_train: np.array, y_train: np.array):

        n = y_train.shape[0]
        b = y_train.reshape(
            -1,
        )
        A = np.hstack([np.ones((n, 1)), x_train.reshape(-1, 1)])

        def loss(w):
            return (
                np.linalg.norm(b - A @ w) ** 2
                + np.sum(np.exp(-100.0 * w)) * (w < 0).any()
            )

        def loss_unconstrained(w):
            return np.linalg.norm(b - A @ w) ** 2

        self.res_ = powell_optimizer(
            loss,
            np.zeros((2,)),
        )
        self.res_unconstrained_ = powell_optimizer(loss_unconstrained, np.zeros((2,)))

        self.coefs_ = self.res_["x"]
        self.coefs_unconstrained_ = self.res_unconstrained_["x"]

    def predict(self, x_test: np.array, mode: str = "default") -> np.array:

        if (not self.coefs_ is None) and (not self.coefs_unconstrained_ is None):
            if mode == "default":
                return self.coefs_[0] + self.coefs_[1] * x_test
            elif mode == "unconstrained":
                return (
                    self.coefs_unconstrained_[0] + self.coefs_unconstrained_[1] * x_test
                )
            else:
                raise ValueError(
                    f"Incorrect mode set. Expected 'default' or 'unconstrained'. Got {mode}"
                )
        else:
            raise ValueError(f"Model is unavailable for predictions.")
```

```python
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]
diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]
```

```python
regr = nnls_custom()
regr.fit(diabetes_X_train, diabetes_y_train)
```
